5. Using a subquery, find the names of all the tracks for the album "Californication".

```python
%sql select * from tracks join albums a on a.id = tracks.album_id where tracks.album_id in (select id from albums where title = 'Californication')
```

6. Find the total number of invoices for each customer along with the customer's full name, city and email.

```python
%sql select first_name, last_name, city, email, count(i.total) from customers inner join invoices i on customers.id = i.customer_id group by customers.id, first_name, last_name, city, email order by customers.id
```

9. What is the song title of trackID 12 from the "For Those About to Rock We Salute You" ?

```python
%sql select name from tracks join albums a on a.id = tracks.album_id where album_id in (select id from albums where title = 'For Those About To Rock We Salute You') and tracks.id = 12
```

10. Retrieve a list with the managers last name, and the last name of the employees who report to him or her. **Who are the reports for the manager named Mitchell?**

```python
%sql select * from employees where reports_to in (select id from employees where last_name = 'Mitchell')
```

11. Find the name and ID of the artists who do not have albums.

12. Use a UNION to create a list of all the employee's and customer's first names and last names ordered by the last name in descending order. **Determine what is the last name of the 6th record** Remember to order things in descending order to be sure to get the correct answer.

```python
%sql select e.first_name, e.last_name from employees e union select c.first_name, c.last_name from customers c order by last_name DESC
```
